```python
import pandas as pd
```

```python
df = pd.read_csv('dat.csv')
df.head()
```

```python
df['scholarship'] = df['scholarship'].map({0: 'No', 1:'Yes'})
df['hypertension'] = df['hypertension'].map({0: 'No', 1:'Yes'})
df['diabetes'] = df['diabetes'].map({0: 'No', 1:'Yes'})
df['alcoholism'] = df['alcoholism'].map({0: 'No', 1:'Yes'})
df['sms_received'] = df['sms_received'].map({0: 'No', 1:'Yes'})
df.head()
```

```python
df = df[['patientID', 'apptID', 'gender', 'age', 'neighborhood', 'scholarship', 'hypertension', 'diabetes', 'alcoholism', 'sched_date', 'appt_date', 'waiting', 'sms_received', 'handicap', 'noshow']]
df.head()
```

```python
df.to_csv('data.csv', index=False)
```
